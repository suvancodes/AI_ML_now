## WSGI
### 1.WSGI = Web Server Gateway Interface
### 2.It’s a standard interface between web servers (like Gunicorn, uWSGI, or mod_wsgi in Apache) and Python web applications (like Flask, Django, FastAPI).
### 3.Flask itself is a WSGI application.

## Jinja2
### Jinja2 = Flask’s template engine.
### It lets you write dynamic HTML with Python-like expressions.
