# convert json to string

```python
import pandas as pd
from io import StringIO as si

# Sample JSON string
data = '''
[
  {"id": 1, "name": "Ankit Roy", "age": 21, "department": "CSE", "cgpa": 8.5},
  {"id": 2, "name": "Sneha Das", "age": 20, "department": "ECE", "cgpa": 9.1},
  {"id": 3, "name": "Rahul Sen", "age": 22, "department": "ME", "cgpa": 7.8},
  {"id": 4, "name": "Priya Sharma", "age": 21, "department": "IT", "cgpa": 8.9}
]
'''
# Convert string to file-like object and read
df = pd.read_json(si(data))
print(df)
print(type(df))
```

# string to json

```python
df.to_json()
```

# using orintation

```python
df.to_json(orient='records')
```

```python
# it is defult
df.to_json(orient='index')
```

# excel file

## make an excel file and read it

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [24, 27, 22],
    'City': ['New York', 'San Francisco', 'Chicago']
}
df1 = pd.DataFrame(data)
df1.to_excel('my_data.xlsx', index=False)
# read it 
df1 = pd.read_excel('my_data.xlsx')
df1
```

## convert this in a pickle file

```python
df1.to_pickle('df1')
```

## read this file

```python
pd.read_pickle('df1')
```
