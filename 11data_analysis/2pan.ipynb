import pandas as pd

# Series ->
### A Series is a one-dimensional array-like object in pandas that can hold data of any type and has labels(called index) associated with each element.

## create a series by list 


```python
data = [1,2,3,4,5]
series = pd.Series(data)
series
```

## create a series by dict

```python
data = {'a':1,'b':2,'c':3}
series = pd.Series(data)
print(series)
```

```python
# create own index 

data = [1,2,3]
ind = ['a','b','c']
series = pd.Series(data,ind)
print(series)
```

```python

                # get ave mark
                
import pandas as pd
import numpy as np

data = {
    'Name': ['Ram', 'Sita', 'Laxman', 'Hanuman'],
    'Maths': [82, 95, 70, 88],
    'Physics': [87, 85, 80, 90],
    'Chemistry': [98, 78, 85, 88]
}

df = pd.DataFrame(data)

df['total'] = df[['Maths','Physics','Chemistry']].sum(axis=1)
df['Average'] = np.round(df['total']/3,2)
df['Grade'] = df['Average'].apply(lambda x: 'A' if x >= 85 else 'B' if x >= 70 else 'C')

topper = df.loc[df['total'].idxmax()]
print('topper is: ',topper['Name'])

print(df)

df.to_csv('result.csv',index=True)

```
