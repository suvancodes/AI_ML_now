### fill value by mean

```python
df = pd.read_csv('student.csv')
df.isnull().any()
```

```python
df['prise_missing'] = df['Price'].fillna(df['Price'].mean())
df
```

## rename the column

```python
df = df.rename(columns={'Name':'Product_Name'})         # columns = {'real_name':'the_name_you_want_to_give'}
df
```

## change the data type

```python
df.dtypes
```

```python
# convart price float to object
df['Price'] = df['Price'].astype(str)
df.dtypes
```

## apply function in column

```python
df['Price'] = df['Price'].astype(float)
df['New_Price'] = df['Price'].apply(lambda x:x*2)
df
```

## data aggregating and grouing

```python
group_mean = df.groupby('Category')['Price'].mean()
group_mean
```

```python
group_sum = df.groupby(['Category','Name'])['prise_missing'].sum()
group_sum
```

## multiple aggregate

```python
group_agg = df.groupby('Category')['prise_missing'].agg(['sum','mean','count'])
group_agg
```

## merging and joining

```python
data1 = {
    'key':['A','B','C'],
    'value1':[1,2,3]
}

data2 = {
    'key':['A','B','F'],
    'value2':[4,5,6]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

pd.merge(df1,df2,on='key',how='inner')
```

```python
pd.merge(df1,df2,on='key',how='outer')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```
