# SQLite

## connect to a data base

```python
import sqlite3 as sql
```

```python

con = sql.connect('example.db')
con
```

## create a cursor

```python
cursor = con.cursor()
```


## Create a table (example)

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS users (
    id INTEGER,
    name TEXT,
    age INTEGER
)''')
con.commit()
```

```python
cursor.execute('''
Select * from users
''')

```

## insert value

```python
cursor.execute('''
Insert into users(id,name,age)
values(123,'subho',19)
''')
cursor.execute('''
Insert into users(id,name,age)
values(1232,'abc',19)
''')
cursor.execute('''
Insert into users(id,name,age)
values(121,'ankit',20)
''')
cursor.execute('''
Insert into users(id,name,age)
values(113,'athar',20)
''')
cursor.execute('''
Insert into users(id,name,age)
values(103,'hello',19)
''')
con.commit()
```

### update data in a table

```python
cursor.execute('''
Update users
set age = 20
where name = "subho"
               ''')
con.commit()
```

### query data form the table

```python
cursor.execute('Select * from users')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

### delete data from table

```python
cursor.execute('''
Delete from users
where name='abc'
               ''')
con.commit()

cursor.execute('Select * from users')
rows = cursor.fetchall()
rows
```
