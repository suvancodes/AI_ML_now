### Reference counting
Reference counting is a memory management technique used by Python to keep track of the number of references pointing to an object in memory. When the reference count drops to zero, it means no variable is using the object anymore, so Python automatically deletes it and frees the memory.

```python
import sys
a = []
print(sys.getrefcount(a))
# the output is 2 beacuse 1 reference coming form vairable a and other coming form getrefcount() this function
b = a

print(sys.getrefcount(b))
# the output is 3 beacuse 1 reference coming form vairable a , 1 coming from b and other coming form getrefcount() this function
```


#### Garbage Collection in Python ðŸ”„ðŸ§¹
Garbage Collection (GC) in Python is the process of automatically freeing memory by deleting objects that are no longer in use, so your program doesn't use up more memory than it need

```python
import gc
# enable the garbage collection
gc.enable()

# disable the garbage collection
gc.disable()
# collect the garbage
gc.collect()

## garbage collection stats
gc.get_stats()

# get unreachable object
gc.garbage
```
