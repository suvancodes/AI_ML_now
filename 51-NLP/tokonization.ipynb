```python
import nltk
import pandas as pd
```

```python
corpus = """ 
hello how are you
I am fine thank you
what about you
I am doing well
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
## Tokenization
## paragraph -> sentence
from nltk.tokenize import sent_tokenize
```

```python
docoment = sent_tokenize(corpus)
```

```python
# paragraph to words
## sentence to words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sen in docoment:
    print(word_tokenize(sen))
```
