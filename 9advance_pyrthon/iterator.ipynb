```python
my_list1 = [1,2,3,4,5,6,7,7,8,8,8,8]

Iterators2 = iter(my_list1)
try:
    print(next(Iterators2))
    print(next(Iterators2))
    print(next(Iterators2))
    print(next(Iterators2))
    print(next(Iterators2))
except StopIteration:
    print('there are no element')
    
```
