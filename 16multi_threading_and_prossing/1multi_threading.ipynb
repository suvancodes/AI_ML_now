## using normal code

```python
import threading
import time

def printl():
    for letter in 'abcd':
        time.sleep(2)
        print(letter)

def printn():
    for num in '12345':
        time.sleep(2)
        print(num)

t = time.time()
printl()
printn()

et = time.time()-t


print(f"excuition time is {et}")
```

## using threading

```python
import threading
import time

def printl():
    for letter in 'abcd':
        time.sleep(2)
        print(letter)

def printn():
    for num in '12345':
        time.sleep(2)
        print(num)
        
### creating two threads        

t1 = threading.Thread(target=printl)
t2 = threading.Thread(target=printn)

t = time.time()

### start two threads

t1.start()
t2.start()

### wait for the thread to complete
t1.join()
t2.join()

et = time.time()-t

print(f"excuition time is {et}")
```
