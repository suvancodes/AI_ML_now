## log massege with different levels

## ðŸ”¢ Python Logging Levels
### Level	Value	Description

### DEBUG	10	Detailed information, typically useful for diagnosing problems. (e.g., variables, flow, internal state)

### INFO	20	General information about program execution, like successful completion of tasks.

### WARNING	30	Something unexpected happened, but the program is still running.

### ERROR	40	A more serious problem that caused a failure in a part of the program.

### CRITICAL	50	A severe error that may prevent the program from continuing to run.



## configer the logging

```python
import logging as log

# Remove all handlers associated with the root logger
for handler in log.root.handlers[:]:
    log.root.removeHandler(handler)

# Now configure logging again
log.basicConfig(
    filename='app.log',
    filemode='w',
    level=log.DEBUG,
    format="%(asctime)s => %(levelname)s => %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)

log.debug('it is a debug message')
log.info('it is a info message')
log.warning('it is a warning message')
log.error('it is a error message')
log.critical('it is a critical message')
```

## some example of logging


```python
import logging as log
logger1 = log.getLogger('module1') 
logger1.setLevel(log.DEBUG)

logger2 = log.getLogger("module2")
logger2.setLevel(log.WARNING)
log.basicConfig(
    
    level=log.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)
```
