## Movie Revenue Prediction

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mlt

url = "https://tmdb-movie-eda.netlify.app/"
df_list = pd.read_html(url)
df = df_list[0]
df
```

### data cleaning

```python
df.isnull().sum()
```

### Budget vs Revenue

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.scatterplot(x='budget', y='revenue', data=df)
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.title("Budget vs Revenue")
plt.show()
```

### top5 earning movie

```python
no = df.groupby('original_title')['revenue'].count()
no_of_show = no.tail(10)
no_of_show
```
