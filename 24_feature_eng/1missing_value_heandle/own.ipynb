```python
import seaborn as sns 
```

```python
df =  sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
sns.histplot(df['age'],kde=True)
df['age'].value_counts()
df['embarked'].value_counts()
```

```python
sns.lineplot(df['age'].value_counts())
```

```python
mean_age = int(df['age'].mean())
mean_age
```

```python
df['mean_age'] = df['age'].fillna(mean_age)
```

```python
df[['age','mean_age']]
```

```python
m_v = df['age'].median()
df['medium_age'] = df['age'].fillna(m_v)
```

```python
df[['age','mean_age','medium_age']]
```

```python
df.head()
```

```python
not_nan = df[df['deck'].notna()]['deck']
mode_deck = not_nan.mode()[0]
mode_deck
```

```python
df['mode_deck'] = df['deck'].fillna(mode_deck)
```

```python
df[['deck','mode_deck']]
```

```python
sns.histplot(df['age'],kde=True)
```
