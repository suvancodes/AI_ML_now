```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from io import StringIO
import pandas as pd
import matplotlib.pyplot as plt
import time
# options = Options()
# options.add_argument("--headless")
# driver = webdriver.Chrome(options=options)


# driver.get('https://terminal-troops.super.site/leaderboard')
# time.sleep(5)  

# html = driver.page_source
# driver.quit()


tables = pd.read_html('https://terminal-troops.super.site/leaderboard')

df = tables[0]
df.columns = df.iloc[0]        
df = df.drop(index=0)          
df = df.reset_index(drop=True)
df.columns = df.columns.str.strip()  
df


```

```python
df['Total points'] = df['Total points'] = pd.to_numeric(df['Total points'], errors='coerce').fillna(0).astype(int)

grp = df.groupby('Member Name')['Total points'].sum()
plt.figure(figsize=(11, 7))
grp.plot(kind='bar', color='green')

plt.xlabel('Member Name')
plt.ylabel('Total Points')
plt.title('Leaderboard Ranking')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('rank.pdf')
```
